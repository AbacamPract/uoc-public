```python
import pandas as pd
import requests
# import json
import os
url = 'https://catalogodatos.cnmc.es/dataset/8a2461de-07b9-4bbc-9e56-db06270f888a/resource/e63f5db2-0433-4b72-8647-e5a4c74f1876/download/ds_14042_1.json'
response = requests.get(url)

# FORZAMOS LA CODIFICACIÓN PARA IGNORAR EL BOM
response.encoding = 'utf-8-sig'
data = response.json()
df = pd.DataFrame(data)
df.info()
```

```python
# df['fecha'] = pd.to_datetime(df['fecha_precio'])
df = df.rename(columns={'promedio_de_pvp_diario_cubo': 'precio'})
df = df.rename(columns={'fecha_precio': 'fecha'})

columnas_interes = [
            'fecha', 
            'provincia', 
            'producto', 
            'precio'
        ]
df_final = df[columnas_interes]
```

```python
# 1. Definir la ruta y el nombre del archivo
relative_path = os.path.join('..', 'src', 'assets', 'data')
file_name = 'Horizont_chart_price_evo_2026.json'
full_path = os.path.join(relative_path, file_name)

os.makedirs(relative_path, exist_ok=True)

# 2. Guardar usando el método nativo de Pandas
df_final.to_json(
    full_path, 
    orient='records',    # Formato [{...}, {...}] ideal para D3.js
    force_ascii=False,   # Mantiene los acentos como 'ó'
    indent=4,            # Lo hace legible
    date_format='iso'    # Asegura que las fechas se vean como "2026-03-14"
)

# print(f"Archivo guardado correctamente en: {full_path}"
```
